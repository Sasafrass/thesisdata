```python
# Set inclusion column to be all zeros
fout['included'] = 0
```

```python
# Set all correct inclusions to 1 in fout['included']
for title in goedincl['title']:
    fout.loc[fout['title'] == title, 'included'] = 1
```

```python
# find duplicates
for i in range(len(titles)):
    for j in range(i+1, len(titles)):
        if titles[i] == titles[j]:
            print(titles[i])
```
